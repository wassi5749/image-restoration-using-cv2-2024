# Image Restoration

```python
 #import useful lbraries that we will use to perform all of these functons to restore the image
    
import cv2                                       # Importing the OpenCV library for the image processing
import numpy as np                               # Then we Import NumPy library or module  for numerical operations
from matplotlib import pyplot as plt             # Importing matplotlib for image visualization
```

```python
                                                 # firstly load the image that is totally damaged
dam_image = cv2.imread('damage image.jpg')       # that function will read the  image and store it in the varable dam_image
```

```python
                                                 # Check if the image is loaded successfully
if dam_image is None:                            # apply if condtion to check the path of the image 
    print("Error: Unable to load the image.")
    exit()                                       # Exit the program if image loading fails
```

```python
                                                         # Display the original image
plt.imshow(cv2.cvtColor(dam_image, cv2.COLOR_BGR2RGB))   # Displaying the original image in color
plt.title('Damage Image')                                # Setting title for the subplot
plt.axis('off')                                          # Turning off axis for better visualization
plt.show()
```

```python
                                                         # Converting the image to grayscale to simplify processing
gray = cv2.cvtColor(dam_image, cv2.COLOR_BGR2GRAY)       # for that we ue cvtcolor functionn
```

```python
                                                # Display the grayscale image 
plt.imshow(gray, cmap='gray')                   # Displaying the grayscale image
plt.title('Grayscale Image')                    # Setting title for the subplot
plt.axis('off')                                 # Turning off axis for better visualization
plt.show()
```

```python
# Apply median filtering to remove noise (means undersired thing) in the image 
# Median filtering helps in reducing noise and preserving edges
# A kernel size of 5x5 is chosen for filtering

median_filter_image= cv2.medianBlur(gray, 5)       #cv2.medianBlur() is a function for applying median filtering to an image
```

```python
                                                  # Display the median-filtered image
plt.imshow(median_filter_image, cmap='gray')      # Displaying the filtered image (after median blur)
plt.title('Filtered Image (Median Blur)')         # Setting title for the subplot
plt.axis('off')                                   # Turning off axis for better visualization
plt.show()
```

```python
# Perform morphological opening to remove thin lines
# Morphological opening is used to remove thin lines or small objects
# A 5x5 kernel of ones is created for the opening operation

kernel5X5 = np.ones((5,5),np.uint8)  
Morph_image= cv2.morphologyEx(median_filter_image, cv2.MORPH_OPEN, kernel5X5)  
```

```python
                                           # Display the result using Matplotlib
plt.imshow(Morph_image, cmap='gray')       # Displaying the image after morphological opening
plt.title(' Morphological Opening')        # Setting title for the subplot
plt.axis('off')                            # Turning off axis for better visualization
plt.show()
```

```python
Restored_image = Morph_image
```

```python
plt.imshow(cv2.cvtColor(Restored_image, cv2.COLOR_BGR2RGB))           #show the restored image in RGB
plt.title('Restored Image')                                           # Setting title for the subplot
plt.axis('off')                                                  
plt.show()
```
